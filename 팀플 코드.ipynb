```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import warnings
warnings.filterwarnings('ignore')
plt.rc('font', family='Malgun Gothic')
```

```python
import win32com.client
 
# 연결 여부 체크
objCpCybos = win32com.client.Dispatch("CpUtil.CpCybos")
bConnect = objCpCybos.IsConnect
if (bConnect == 0):
    print("PLUS가 정상적으로 연결되지 않음. ")
    exit()
```

```python
codemgr = win32com.client.Dispatch("CpUtil.CpCodeMgr")

kospi_code = codemgr.GetStockListByMarket(1) # 1: 코스피
kosdaq_code = codemgr.GetStockListByMarket(2) # 2: 코스닥
```

```python
# 리스트로 형변환
kospi_code = list(kospi_code)
kosdaq_code = list(kosdaq_code)
# CpStockCode를 사용하여 종목 코드를 종목명으로 나타내기
stock_code = win32com.client.Dispatch('CpUtil.CpStockCode')

kospi_name = [stock_code.CodeToName(code) for code in kospi_code]
kosdaq_name = [stock_code.CodeToName(code) for code in kosdaq_code]
```

```python
code_list = []
name_list = []
day_list = []
time_list = []
start_list = []
high_list = []
low_list = []
end_list = []
volumn_list = []
```

```python
# 코스피 주식데이터 추출
from tqdm import tqdm_notebook

stockchart = win32com.client.Dispatch('CpSysDib.StockChart')

for code, name in tqdm_notebook(zip(kospi_code, kospi_name)):
    stockchart.SetInputValue(0, code)
    stockchart.SetInputValue(1, ord('1'))
    stockchart.SetInputValue(3, 20220428)
    stockchart.SetInputValue(5, [0, 1, 2, 3, 4, 5, 8])
    stockchart.SetInputValue(6, ord('D'))
    stockchart.SetInputValue(9, ord('1'))
    stockchart.BlockRequest()

    numData = stockchart.GetHeaderValue(3)

    for i in range(numData):
        code_list.append(code)
        name_list.append(name)
        day_list.append(stockchart.GetDataValue(0, i))
        time_list.append(stockchart.GetDataValue(1, i))
        start_list.append(stockchart.GetDataValue(2, i))
        high_list.append(stockchart.GetDataValue(3, i))
        low_list.append(stockchart.GetDataValue(4, i))
        end_list.append(stockchart.GetDataValue(5, i))
        volumn_list.append(stockchart.GetDataValue(6, i))

    time.sleep(0.28)
```

```python
# 딕셔너리 생성
columns = ['종목코드', '회사명', '날짜', '시가', '고가', '저가', '종가', '거래량']
kospi_dict = {
      '종목코드' : code_list, '회사명' : name_list , '날짜' : day_list,  '시가' : start_list
    , '고가' : high_list, '저가' : low_list, '종가' : end_list, '거래량' : volumn_list
}
```

```python
# 데이터 프레임 생성
kospi_stock_data = pd.DataFrame(
      kospi_dict
    , columns = columns
)
kospi_stock_data.head()
```

```python
# 데이터 프레임 저장
kospi_stock_data.to_csv('kospi_stock_data.csv', encoding = 'utf-8-sig')
```

```python
# 분봉 데이터 뽑기
from tqdm import tqdm


result = []

def stock_data_extract(corpCode, count, dwn):

    stockchart = win32com.client.Dispatch('CpSysDib.StockChart')

    for code in tqdm(corpCode):
        stockchart.SetInputValue(0, code)
        stockchart.SetInputValue(1, ord('2'))
        stockchart.SetInputValue(4, count)
        stockchart.SetInputValue(5, [0, 1, 2, 3, 4, 5, 8, 9, 20])
        stockchart.SetInputValue(6, ord(dwn))
        stockchart.SetInputValue(9, ord('1'))


        stockchart.BlockRequest()
        numData = stockchart.GetHeaderValue(3)
        time.sleep(0.28)

        for i in range(numData):
            record = []
            
            record.append(stockchart.GetDataValue(0, i)) # 날짜
            record.append(stockchart.GetDataValue(1, i)) # 시간
            record.append(stockchart.GetDataValue(2, i)) # 시가
            record.append(stockchart.GetDataValue(3, i)) # 고가
            record.append(stockchart.GetDataValue(4, i)) # 저가
            record.append(stockchart.GetDataValue(5, i)) # 종가
            record.append(stockchart.GetDataValue(6, i)) # 거래량
            record.append(stockchart.GetDataValue(7, i)) # 거래대금
            record.append(stockchart.GetDataValue(8, i)) # 기관순매수
            result.append(record)
            

    return True
```

```python
min_stock_data = stock_data_extract(corpCode, count, dwn)
if min_stock_data == False:
    exit()

NextCount = 1
while stockchart:  #연속 조회처리
    NextCount+=1
    if (NextCount > 1):
        break
    min_stock_data = stock_data_extract(corpCode, count, dwn)
    if min_stock_data == False:
        exit()
```

```python
# 데이터 프레임 생성=> xlsx 저장
import xlsxwriter
f = 'stock_data_카카오_m.xlsx'
stock_data = pd.DataFrame(
      result
    , columns = columns
)
writer = pd.ExcelWriter(f, engine='xlsxwriter')
stock_data.to_excel(writer)
writer.save()
```
